# Faceted cubesSplit by a scenario coordinate using `facet_wrap`.Panels share the same fill scale and legend.

```python
import numpy as np
import xarray as xr
from cubedynamics.plotting.cube_plot import CubePlot

scenarios = ['baseline', 'mitigation', 'adaptation']
data = xr.DataArray(np.random.random((3, 4, 6, 6)), dims=('scenario','time','y','x'), coords={'scenario': scenarios}, name='precip')
plot = CubePlot(data, show_progress=False, title='Scenario comparison', caption={'text': 'Panels share color mapping.'})
plot.facet_wrap('scenario', ncol=2)
html = plot.to_html()
plot.save('facet_example.html')
html
```
