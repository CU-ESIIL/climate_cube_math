# Annotations exampleAdd planes and text labels to call out key features.

```python
import numpy as np
import xarray as xr
from cubedynamics.plotting.cube_plot import CubePlot

data = xr.DataArray(np.linspace(0, 1, 64).reshape(4,4,4), dims=('time','y','x'), name='density')
plot = CubePlot(data, show_progress=False, title='Annotated cube', caption={'text': 'Example of annotation helpers.'})
plot.annot_plane('time', 2, text='Event threshold')
plot.annot_text((1,1,1), 'Center point')
html = plot.to_html()
html
```
