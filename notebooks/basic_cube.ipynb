# Basic cube viewerUse CubePlot via the `v.plot` verb to preview a simple cube.

```python
import numpy as np
import xarray as xr
import cubedynamics.verbs as v

# toy cube
data = xr.DataArray(np.random.random((4, 8, 8)), dims=('time','y','x'), name='value')

# pipe-style rendering
_ = v.plot(title='Basic cube', legend_title='Value', fig_text='Simple random cube')(data)
```
