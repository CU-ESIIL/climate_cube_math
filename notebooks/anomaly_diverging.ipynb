# Time anomaly with diverging paletteCompute a z-score anomaly over time and visualize with a diverging palette.

```python
import numpy as np
import xarray as xr
from cubedynamics.plotting.cube_plot import CubePlot, ScaleFillContinuous, CubeAes, CubeLayer
from cubedynamics.plotting.cube_plot import stat_time_anomaly, geom_cube

data = xr.DataArray(np.random.randn(6, 6, 6), dims=('time','y','x'), name='temp')
plot = CubePlot(data, show_progress=False, title='Time anomaly', caption={'text': 'Z-score anomaly relative to time mean.'})
plot.layers = [CubeLayer(geom='cube', stat='time_anomaly', aes=CubeAes())]
plot.fill_scale = ScaleFillContinuous(center=0.0, palette='diverging', name='Anomaly (z)')
plot_html = plot.to_html()
plot.save('anomaly_example.html')
plot_html
```
